```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy.special import comb 
```

## Initialize Dataset \& Weights

```python
def init_(p, n):
    
    """ Initializes system parameters.
    
    :param p: The number of training sample
    :param n: The number features (dimensions)
    :returns:
        x: The training sample
        s: The true labels of training sample
        w: The weight vector      
    """

    # Generate dataset with samples that follow Gaussian distribution
    mu = np.zeros(p) 
    sigma = np.ones(p)
    # No correlation between features
    cov = np.eye(p)
    x = np.random.multivariate_normal(mu, cov, n).T

    # For the P points, randomly assign a label (either +1 or -1) with equal probability
    s = random.choices( [1, -1], weights=[1/2, 1/2], k=p )
    s = np.array(s)

    # Initialize weights for a single neuron and each feature
    w = np.zeros(n)

    return x, s, w
```

## Plot the Probability of being Linearly Separable 

```python
def successfulPlot(A, prob, label):
    
    """ Plots the fraction Pls = C(P, N)/2^P of linearly separable functions versus a = P/N
    
    :param A: The ratio of P to N
    :param prob: The probability of being linearly separable 

    """
    
    plt.suptitle("The fraction Pls = C(P, N)/2^P of linearly separable functions versus Î± = P/N", fontsize=14,\
                 fontweight="bold")
    plt.xlabel("a = P/N")
    plt.ylabel("C(P,N) / 2^P")
    plt.plot(A, prob, label=label)
```

## Rosenblatt Perceptron Algorithm Implementation

```python
def rosenblattPerceptron(x, s, w, P, c): 
    
    """ Implements Rosenblatt-Perceptron algorithm
    
    :param x: The training sample
    :param s: The true labels of training sample
    :param w: The weight vector
    :param P: The number of training samples
    :param c: The margin of whether weights change

    """
    
    for i in range(epochs):

        E = np.array([])
        
        for j in range(P):
            # Predict the labels
            E = np.append(E, x[j].dot(w.T)*s[j])

            # Update weights in case predicted value != true value
            if (E[j] <= c):
                w = w + (x[j] * s[j])/N

        # Stop iterating when all predicted values equal true values
        if (np.all(E > c)):
            print("All instances are predicted correctly")
            break
            
    print("The accuracy of the model for N =", +N, "and P =", +P, "equals:", +np.mean(E > c), "\n")
    
    return np.mean(E > c)
```

## The Capacity of a Hyperplane

```python
# Number of features (columns)
N = 40
# The different ratios of P to N
A = np.arange(0.75, 3.25, 0.25)
# The margin of whether weights change
c = 0
# Initialize number of epochs
epochs = 50
# Number of repeated datasets
n = 50
prob = np.array([])

for a in A:
    accuracy = np.array([])
    # Number of training samples (rows)
    P = int(a*N) 
    
    # Generate n independent datasets
    for j in range(n):
        # Initialize training samples given certain N and P 
        x, s, w = init_(P, N)
        # Run Rosenblatt-Perceptron algorithm
        accuracy = np.append(accuracy, rosenblattPerceptron(x, s, w, P, c) )
        
    prob = np.append(prob, np.sum(accuracy[accuracy==1])/n)

plt.figure(figsize=(8, 5), dpi=80)
successfulPlot(A, prob, 'Probability of successful runs')
plt.legend(loc='best')
plt.show()
```

### Extension-2

Determine the embedding strenghts $\chi^\mu$ (or formulate the algorithm in terms of the $\chi^\mu$) to
obtain a histogram which reflects their frequency in the successful cases upon convergence
of the training.

```python
def rosenblattPerceptronEmb(x, s, w, P, c): 
    
    """ Implements Rosenblatt-Perceptron algorithm with updates of the embedding strengths
    
    :param x: The training sample
    :param s: The true labels of training sample
    :param w: The weight vector
    :param P: The number of training samples
    :param c: The margin of whether weights change

    """
    
    # Array of the embedding strengths (initialization outside of the epochs loop, right?)
    es = [0]*P

        
    for i in range(epochs):

        E = np.array([])
        
        for j in range(P):
            # Predict the labels
            E = np.append(E, x[j].dot(w.T)*s[j])

            # Update weights and increase corresponding embedding strength in case predicted value != true value
            if (E[j] <= c):
                w = w + (x[j] * s[j])/N
                es[j] = es[j] + 1

        # Stop iterating when all predicted values equal true values
        if (np.all(E > c)):
            print("All instances are predicted correctly")
            break
            
    print("The accuracy of the model for N =", +N, "and P =", +P, "equals:", +np.mean(E > c))
    print("Embedding strengths:",es, "\n")
    return np.mean(E > c),es
```

```python
# Number of features (columns)
N = 20
# The different ratios of P to N
A = np.arange(0.75, 3.25, 0.25)
# The margin of whether weights change
c = 0
# Initialize number of epochs
epochs = 50
# Number of repeated datasets
n = 1
prob = np.array([])
# Initialize array of embedding strengths
emb_strengths = []

for a in A:
    # Initialize array of accuracies
    accuracy = np.array([])
    # Number of training samples (rows)
    P = int(a*N) 
    
    # Generate n independent datasets
    for j in range(n):
        # Initialize training samples given certain N and P 
        x, s, w = init_(P, N)
        # Run Rosenblatt-Perceptron algorithm
        acc,es = rosenblattPerceptronEmb(x, s, w, P, c)
        accuracy = np.append(accuracy, acc)
        emb_strengths.append(es)
        
    prob = np.append(prob, np.sum(accuracy[accuracy==1])/n)

# Plot histogram of embedding strenths of successful runs
for i in range(len(prob)):
    if prob[i] == 1:
        plt.suptitle("Histogram of embedding strenths")
        plt.xlabel("Number of updates")
        plt.ylabel("Frequency")
        plt.hist(emb_strengths[i])
#         plt.savefig('emb_str'+ str(i))
#         plt.clf()
        plt.show()
        
    
```
